```python


import nibabel as nib
import matplotlib.pyplot as plt

def load_and_plot_2d_nii(file_path):
    # Load the .nii or .nii.gz file
    img = nib.load(file_path)
    data = img.get_fdata()
    
    # Check if data is 2D
    if data.ndim != 2:
        print("The loaded data is not 2D.")
        return
    
    # Plot the image
    plt.imshow(data, cmap="gray")
    plt.axis("off")
    plt.show()

# Example usage
main_path= "/Users/saschastocker/Documents/Stanford/work2024/WholeHeartCropISMRM/Summer_of_Segmentation_2024/Le/Volunteer_04/Distortion_Corrected/DiVO_15_04"
file_paths =[ "03_Segmentation_Images/Average_Diffusion_Weighted_Image_Slice_002.nii", "05_Segmentation_Images_CI/Cropped_Average_Diffusion_Weighted_Image_Slice_002.nii", "05_Segmentation_Images_CI/Cropped_Mean_Diffusivty_Image_Slice_002.nii"]
for file in file_paths:
    load_and_plot_2d_nii(main_path+ '/'+ file)
```
